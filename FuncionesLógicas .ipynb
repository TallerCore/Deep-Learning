
# Modelos McCulloch-Pitts

```python
import numpy as np

def unit_step(v):
	""" Heavyside Step function. v must be a scalar """
	if v >= 0:
		return 1
	else:
		return 0
	
def perceptron(x, w, b):
  """ Function implemented by a perceptron with weight vector w and bias b """
  v = np.dot(w, x) + b
  y = unit_step(v)
  return y, w, b
```

```python
def LogicalNOT(x):
  return perceptron(x, w=-1, b=0.5)
```

```python
y1, w, b = LogicalNOT(0)
y2, w, b = LogicalNOT(1)

print("NOT(0) = {}, peso = {}, bias = {}".format(y1,w,b))
print("NOT(1) = {}, peso = {}, bias = {}".format(y2,w,b))
```

```python
def LogicalAND(x):
    w = np.array([1, 1])
    b = -1.5
    return perceptron(x, w, b)
```

```python
y1, w, b = LogicalAND(np.array([1, 1]))
y2, w, b = LogicalAND(np.array([1, 0]))
y3, w, b = LogicalAND(np.array([1, 1]))
y4, w, b = LogicalAND(np.array([1, 0]))

print("AND({}, {}) = {}, peso = {}, bias = {}".format(1, 1, y1,w,b))
print("AND({}, {}) = {}, peso = {}, bias = {}".format(1, 0, y2,w,b))
print("AND({}, {}) = {}, peso = {}, bias = {}".format(0, 1, y3,w,b))
print("AND({}, {}) = {}, peso = {}, bias = {}".format(0, 0, y4,w,b))
```

```python
def LogicalOR(x):
    w = np.array([1, 1])
    b = -0.5
    return perceptron(x, w, b)
```

```python
y1, w, b = LogicalOR(np.array([1, 1]))
y2, w, b = LogicalOR(np.array([1, 0]))
y3, w, b = LogicalOR(np.array([1, 1]))
y4, w, b = LogicalOR(np.array([1, 0]))

print("AND({}, {}) = {}, peso = {}, bias = {}".format(1, 1, y1,w,b))
print("AND({}, {}) = {}, peso = {}, bias = {}".format(1, 0, y2,w,b))
print("AND({}, {}) = {}, peso = {}, bias = {}".format(0, 1, y3,w,b))
print("AND({}, {}) = {}, peso = {}, bias = {}".format(0, 0, y4,w,b))
```
