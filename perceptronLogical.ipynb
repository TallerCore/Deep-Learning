```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
class Perceptron:
    def __init__(self, n_input, learning_rate):
        self.w = -1 + 2 * np.random.rand(n_input)
        self.b = -1 + 2 * np.random.rand()
        self.eta = learning_rate
    
    def predict(self, X):
        p = X.shape[1]
        y_est = np.zeros(p)
        for i in range (p):
            y_est[i] = np.dot(self.w,X[:,i]) + self.b
            if y_est[i] >=0:
                y_est[i] = 1
            else:
                y_est[i] = 0
        return y_est
    
    def fit(self, X, Y, epochs = 50):
        p = X.shape[1]
        for _ in range(epochs):
            for i in range(p):
                y_est = self.predict(X[:,i].reshape(-1,1))
                self.w += self.eta * (Y[i]-y_est)* X[:,i]
                self.b += self.eta * (Y[i]-y_est)*1
    
def draw(modelo):
    w1, w2, b = modelo.w[0], modelo.w[1], modelo.b
    li, ls = -2, 2
    plt.plot([li,ls],[(1/w2)*(-w1*(li)-b), (1/w2)*(-w1*(ls)-b)], '--r')
```

```python
neuron = Perceptron (2, 0.1)
X = np.array([[0,0,1,1],
              [0,1,0,1]])
Y = np.array([0,0,0,1])
```

```python
neuron.fit(X,Y)
```

```python
neuron.w, neuron.b
```

```python
_, p = X.shape
for i in range(p):
    if Y[i] == 0:
        plt.plot(X[0,i], X[1,i], 'ob')
    else:
        plt.plot(X[0,i], X[1,i], 'og')
plt.title('Perceptron AND')
plt.grid('on')
plt.ylim([-2,2])
plt.xlim([-2,2])
plt.xlabel('X1')
plt.ylabel('X2')
draw(neuron)
plt.show()
```
